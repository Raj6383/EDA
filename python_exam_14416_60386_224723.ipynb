### 1. Describe the difference between univariate, bivariate and multivariate analysis? (8 MARKS)

1) Univariate analysis : This type of data consists of only one variable. The analysis of univariant data is thus the simplest
    form of analysis since the information deals with only one quantity that changes. it does not deal with causes or
    relationship and the main purpose of analysis is to describe the data and find the patterns that exists within it.
    The example of univariante data can be Height.
    Height = [164,167.3,170,174.2,178,180,186]
    Suppose that height of 7 students of a class is recorded, there is only one variable that is height and its not dealing
    with any cause or relationship. the description of patterns found in this type of dya can be made by drawing conclusion using 
    Central tendency measures,Dispersion,Spread of data and by using frequency distribution tables

2) Bivariate analysis : This type of data involves two different variable.The analysis of this type of data deals 
    with causes and relationshipand the analysis is done to find out the relationship among the two variables. 
    Examples of bivariate data can be temprature and icecream sales in summer season.
    Temp=[20,25,35,43]
    icecream_sales[2000,2500,3000,3500]
    
    here the relationship is directly visible from table that temperature and sales are directly proportion to each other
    thus temperature increase sales also increase
    the data analysis involves comparision,relationship,causes and explanation

3) Multivariate analysis : This type of data involves three or more variables, it is categorised under multivariate.
    Example of this data is suppose an advertiser wants to compare the popularity of 4 advertisementon a website.
    then click rates could be measure for both men and women and the relationship between variables can be examined.
    It is similar tu bivariate but contains more than 1 dependent variable 
    ways to perform analysis on this data depends on goals to be achieved . some of the techniques are regression analysis,
    path analysis,factor analysis and Multivariate analysis of variance

#### D. Write a function to normalize the `Length` column within the range [2,3]. (5 marks)"

```python
import numpy as np
import pandas as pd 

def minmaxscaler(x):
    a=list()
    for i in x:
        a.append((x - x.min())/(x.max()-x.min())+2)
    return a


print(np.max(minmaxscaler(df['Length'])))
print(np.min(minmaxscaler(df['Length'])))
```
