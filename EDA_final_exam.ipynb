```python
# 5.D Identify columns with ? and fill with median
for i in adult.columns:
    if np.dtype(adult[i]) =='O':
        adult[i]=adult[i].str.strip()
l = []
for i in adult.columns:
    if '?' in adult[i].unique():
        l.append(i)
```

```python
for i in l:
    adult[i].replace('?',adult[i].mode()[0],inplace=True)
```

```python
adult.head()
```

```python
# 5.E.1 Log tansfromation and visualization
lg = adult['age'].transform(lambda x:np.log(x))
fig = plt.figure(figsize=(14,5))
title = fig.suptitle('Distribution of age',fontsize=14)
ax1=fig.add_subplot(1,2,1)
a = sns.distplot(adult['age'],ax=ax1,kde=True,hist=True)
t1 = ax1.set_title('Distribution of age without tranformation')
ax2= fig.add_subplot(1,2,2)
b = sns.distplot(lg,ax=ax2,hist=True,kde=True)
t2= ax2.set_title('Distribution of age with tranformation')
plt.show()
```

```python
# Log transformation makes the graph tend to normal
```

```python
import scipy.stats as st
```

```python
# 5.E.2 visualization of z score on fnlwgt
z = pd.DataFrame(st.zscore(adult['fnlwgt']))
```

```python
fig = plt.figure(figsize=(14,5))
title = fig.suptitle('Distribution of fnlwgt',fontsize=14)
ax1=fig.add_subplot(1,2,1)
a = sns.distplot(adult['fnlwgt'],ax=ax1,kde=True,hist=True)
t1 = ax1.set_title('Distribution of fnlwgt without tranformation')
ax2= fig.add_subplot(1,2,2)
b = sns.distplot(z,ax=ax2,hist=True,kde=True)
t2= ax2.set_title('Distribution of fnlwgt with tranformation')
plt.show()
```
