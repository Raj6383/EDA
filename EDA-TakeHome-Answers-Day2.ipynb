####  1.7 importing import skew, norm

```python
from scipy.stats import skew, norm, probplot, boxcox, f_oneway
```

#### 1.8 Create the numeric_features to remove skewness

```python
numeric_features = list(data.dtypes[(data.dtypes != "str") & (data.dtypes !='object')].index) 
```

#### 1.9 Create the skewed_features = wines[numeric_features]

```python
skewed_features = data[numeric_features].apply(lambda x : skew (x.dropna())).sort_values(ascending=False)
```

#### 1.10 Create DataFrame for skewness?

```python
#compute skewness
skewness = pd.DataFrame({'Skew' :skewed_features})   
```

#### 1.11 Create to Get only higest skewed = [abs(skewness) > 0.7] features?

```python
# Get only higest skewed features
skewness = skewness[abs(skewness) > 0.7]
skewness = skewness.dropna()
print ("There are {} higest skewed numerical features to box cox transform".format(skewness.shape[0]))
```

#### 1.12 Create skewness.index and new skewed_features2

```python
#Create Empty list 
l_opt = {}

#create for loop for skewness.index
for feat in skewness.index:
    data[feat], l_opt[feat] = boxcox((data[feat]+1))
    
#create skewed_features2 for skewness.index
skewed_features2 = data[skewness.index].apply(lambda x : skew (x.dropna())).sort_values(ascending=False)
```

### 1.13 Create new DataFrame for skewed_features2?

```python
#compute skewness
skewness2 = pd.DataFrame({'New Skew' :skewed_features2}) 

# concat both skewness, skewness2
display(pd.concat([skewness, skewness2], axis=1).sort_values(by=['Skew'], ascending=False))
```
